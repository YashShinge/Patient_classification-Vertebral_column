## Classification of patients  - based on Vertebral column  data

> Classifying patients into two categories based on six biomechanical features.  
The two categories are:
>    1. Normal
>    2. Abnormal i.e patients with Disk Hernia or Spondylolisthesis
----

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import pickle
```

```python
#helper funcs
def get_feature_names():
    
    cn = open('../input/feature_names.txt').read()
    cn = cn.split(',')
    cn[:] = cn[:-1] + cn[-1].split('and')
    cn = [i.strip().replace(' ', '_') for i in cn]
    
    return cn
    
    
def plot_data_dist(df):
    
    for i in df.columns[:-1]:
        plt.subplot(1,2,1)
        sns.kdeplot(df[df['target'] == 0][i], label = 'Normal')
        sns.kdeplot(df[df['target'] == 1][i], label = 'Abnormal')
        plt.legend()
        plt.grid('on')
        plt.title(f'{i} distribution by classes')

        plt.subplot(1,2,2)
        sns.stripplot(df['target'], df[i], jitter=True)
        plt.legend(labels=['Normal', 'Abnormal'])
        plt.grid('on')

        plt.tight_layout()
        plt.show()
        
        
def get_outlier_indices(data):
    
    indices = []
    for i in data.columns[:-1]:
        q1 = np.percentile(data[i], 25)
        q3 = np.percentile(data[i], 75)
        iqr = q3 - q1
        step = iqr * 1.5
        indices.extend(data[(data[i] < q1 - step) | (data[i] > q3 + step)].index)
        
    indices = Counter(indices)
    outliers_indices = list(c for c,k in indices.items() if k > 2)
    
    return outliers_indices
        
        
def conf_mat(true, pred):
    
    cmat = metrics.confusion_matrix(true, pred)
    df_cm = pd.DataFrame(cmat, index=['Normal', 'Abnormal'], columns=['Normal', 'Abnormal'])
    plt.figure(figsize = (10,7))
    sns.heatmap(df_cm, annot=True)
    plt.xlabel('Predicted Class')
    plt.ylabel('Actual Class')
    plt.show()    
        

def get_ra_score(test, pred_probs):
    f, t, threshold = metrics.roc_curve(test, pred_probs)
    ra = metrics.auc(f, t)
    return f, t, ra

        
def plot_roc_auc(fpr, tpr, ra):
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label=f'AUC = {round(ra, 3)}')
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.grid('on')
    plt.show()
    

def performance_report(model, f1, recall, roc_auc):
    report = pd.DataFrame({"Model": [model],
                            "F1 score": [f1],
                            "Recall_score": [recall],     
                            "AUC_ROC": [roc_auc]
                            })
    return report
```

----
## Training

```python
X = scaled_df.copy()
y = df_noout['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=9, shuffle=True)
```

### 1. Logistic regression - unweighted

```python
log_reg = LogisticRegression(n_jobs=-1)
```

```python
log_reg.fit(x_train, y_train)
```

```python
preds_lr = log_reg.predict(x_test)
pred_prob_lr = log_reg.predict_proba(x_test)[:,1]
```

```python
rec_lr = metrics.recall_score(y_test, preds_lr)
print(f'Recall score: {rec_lr}')
```

```python
f1_lr = metrics.f1_score(y_test, preds_lr)
print(f'F1-score: {f1_lr}')
```

```python
fpr_lr, tpr_lr, roc_auc_lr = get_ra_score(y_test, pred_prob_lr)
print(f'Area under curve: {roc_auc_lr}')
```

```python
plot_roc_auc(fpr_lr, tpr_lr, roc_auc_lr)
```

```python
conf_mat(y_test, preds_lr)
```

----
### 2. Logistic regression weighted - classes

##### a. Grid search

```python
model_lr = LogisticRegression(solver='lbfgs')

#Setting the range for class weights
weights = np.linspace(0.0, 0.99, 200)

#Creating a dictionary grid for grid search
param_grid = {'class_weight': [{0:x, 1:1.0-x} for x in weights]}

#Fitting grid search to the train data with 5 folds
gridsearch = GridSearchCV(estimator= model_lr, 
                          param_grid= param_grid,
                          cv=StratifiedKFold(), 
                          n_jobs=-1, 
                          scoring='f1',
                          verbose=1).fit(x_train, y_train)
```

```python
gridsearch.best_params_
```

```python
weigh_data = pd.DataFrame({ 'score': gridsearch.cv_results_['mean_test_score'], 'weight': (1- weights)})
sns.lineplot(weigh_data['weight'], weigh_data['score'])
plt.xlabel('Weight for class Abnormal')
plt.ylabel('F1 score')
plt.grid('on')
plt.show()
```

##### b. using the best parameters

```python
log_reg_wc = LogisticRegression(**gridsearch.best_params_, n_jobs=-1)
log_reg_wc.fit(x_train, y_train)
```

```python
preds_lrwc = log_reg_wc.predict(x_test)
pred_prob_lrwc = log_reg_wc.predict_proba(x_test)[:, 1]
```

```python
rec_lrwc = metrics.recall_score(y_test, preds_lrwc)
print(f'Recall score: {rec_lrwc}')
```

```python
f1_lrwc = metrics.f1_score(y_test, preds_lrwc)
print(f'F1-score: {f1_lrwc}')
```

```python
print(metrics.classification_report(y_test, preds_lrwc))
```

```python
fpr_lrwc, tpr_lrwc, roc_auc_lrwc = get_ra_score(y_test, pred_prob_lrwc)
print(f'Area under curve: {roc_auc_lrwc}')
```

```python
plot_roc_auc(fpr_lrwc, tpr_lrwc, roc_auc_lrwc)
```

```python
conf_mat(y_test, preds_lrwc)
```

----
## Comparing the models

```python
logreg_report = performance_report('Log reg unweighted', f1_lr, rec_lr, roc_auc_lr)
logreg_weighted_report = performance_report('Log reg weighted', f1_lrwc, rec_lrwc, roc_auc_lrwc)
```

```python
all_model_performance = pd.concat([logreg_report, logreg_weighted_report], axis = 0)
all_model_performance = all_model_performance.set_index(['Model'])
```

```python
all_model_performance
```

We observe, the logistic regression model with class weights has higher f1 and recall score.   
Thus, we select it as our final model.

```python
# get recall score, f1-score and roc_auc_score
f1_score = metrics.f1_score(y, predictions)
print(f'F1-score: {f1_score}')

recall = metrics.recall_score(y, predictions)
print(f'Recall score: {recall}')

roc_auc = metrics.roc_auc_score(y, prediction_probabilities)
print(f'AUC ROC score: {roc_auc}')
```
